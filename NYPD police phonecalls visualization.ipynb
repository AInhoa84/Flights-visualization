Now that we have the base map of New York, before starting to plot our dataset (as there are almost 400K rows), we will keep only the coordinates which have more than 2 phonecalls to the police. Then, we will try to plot a heatmap by volume of phonecalls per location.

```python
#We first create a new dataframe containing only the coordinates and removing Nans:
coords = req_df[['Latitude','Longitude']].copy()
coords = coords.dropna(subset=['Longitude'])
coords = coords.dropna(subset=['Latitude'])
```

```python
#We manipulate the dataframe to group by each pair of coordinates:
import numpy as np
coords['Volume'] = np.ones(len(coords))
coords.head()
```

```python
#We calculate the volume of phonecalls by each coordinate pair:
volumes = pd.DataFrame(pd.pivot_table(coords,index=["Latitude","Longitude"],values=["Volume"],aggfunc=[sum]))
volumes.reset_index(inplace=True)
volumes.columns = volumes.columns.droplevel(1)
volumes.head()
```

```python
#And now we just keep the highest volumes (phonecalls > 2):
top_volumes = volumes[volumes['sum'] > 2]
top_volumes.head()
```

```python
#Now we are ready to plot our figures over the NY map:
from folium import plugins
from folium.plugins import HeatMap
NY_COORDINATES = (40.7, -74.0)
map = folium.Map(location=NY_COORDINATES,zoom_start=10)
heat_data = [[row['Latitude'],row['Longitude']] for index, row in top_volumes.iterrows()]
HeatMap(heat_data,radius=10).add_to(map)
display(map)
```
